```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
grid = np.zeros((8,8,3))
#help(np.ones)
```

```python
# help(plt.imshow)
white = (1,1,1)
for i in range(8):
    if i%2 == 0:
        for j in range(8):
            if j%2 == 0:
                grid[i][j] = white
    else:
        for j in range(8):
            if j%2 == 1:
                grid[i][j] = white


plt.imshow(grid)
```

```python
class Board:
    """Board class"""
    def __init__(self):
        """Creates blank 8x8 grid chess board"""
        self.grid = np.zeros((8,8,3))
        self.blue = None
        self.red = None

        for i in range(8):
            if i%2 == 0:
                for j in range(8):
                    if j%2 == 0:
                        self. grid[i][j] = (1,1,1)
            else:
                for j in range(8):
                    if j%2 == 1:
                        self.grid[i][j] = (1,1,1)

        plt.imshow(self.grid)

    def add_red(self, row:int, col:int):
        """method that accepts a row and column as input which colors corresponding cell"""
        self.grid[row][col] = (1,.2,0)
        self.red = {
            'row': row,
            'column': col
            }

    def add_blue(self, row:int, col:int):
        """ method that accepts a row and column as input which colors corresponding cell"""
        self.grid[row][col] = (0,1,1)
        self.blue = {
            'row': row,
            'column': col
            }

    def render(self):
        """method that displays the chess board on screen with red and blue shown in correct           locations"""
        plt.imshow(self.grid)

    def is_under_attack(self) -> bool:
        """method that return boolean if red is under attack by a blue piece horizontally,             vertically or diagonally"""
        col_diff = abs(self.red['column'] - self.blue['column'])
        row_diff = abs(self.red['row'] - self.blue['row'])

        if self.red['column'] == self.blue['column']:
            return True
        elif self.red['row'] == self.blue['row']:
            return True
        elif col_diff == row_diff:
            return True

        return False
```

```python
new_board = Board()
new_board.add_red(1,6)
new_board.add_blue(6,6)
new_board.render()
assert new_board.is_under_attack() == True
```

```python
new_board.__init__()
new_board.add_red(2,1)
new_board.add_blue(2,7)
new_board.render()
assert new_board.is_under_attack() == True
```
