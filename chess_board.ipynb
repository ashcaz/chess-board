```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
grid = np.zeros((8,8,3))
#help(np.ones)
```

```python
# help(plt.imshow)
white = (1,1,1)
for i in range(8):
    if i%2 == 0:
        for j in range(8):
            if j%2 == 0:
                grid[i][j] = white
    else:
        for j in range(8):
            if j%2 == 1:
                grid[i][j] = white


plt.imshow(grid)
```

```python
class Board:
    def __init__(self):
        self.grid = np.zeros((8,8,3))
        self.blue = None
        self.red = None

        for i in range(8):
            if i%2 == 0:
                for j in range(8):
                    if j%2 == 0:
                        self. grid[i][j] = (1,1,1)
            else:
                for j in range(8):
                    if j%2 == 1:
                        self.grid[i][j] = (1,1,1)

        plt.imshow(self.grid)

    def add_red(self, row, col):
        self.grid[row][col] = (1,.2,0)
        self.red = {
            'row': row,
            'column': col
            }

    def add_blue(self, row, col):
        self.grid[row][col] = (0,1,1)
        self.blue = {
            'row': row,
            'column': col
            }

    def render(self):
        plt.imshow(self.grid)

    def is_under_attack(self) -> bool:
        col_diff = abs(self.red['column'] - self.blue['column'])
        row_diff = abs(self.red['row'] - self.blue['row'])

        print(row_diff, col_diff)

        if self.red['column'] == self.blue['column']:
            return True
        elif self.red['row'] == self.blue['row']:
            return True
        elif col_diff == row_diff:
            return True

        return False
```

```python
new_board = Board()
new_board.add_red(1,6)
new_board.add_blue(6,6)
new_board.render()
print(new_board.is_under_attack())
```

```python
new_board.__init__()
new_board.add_red(2,1)
new_board.add_blue(2,7)
new_board.render()
print(new_board.is_under_attack())
```
